```python
import requests
from bs4 import BeautifulSoup
from itertools import izip
import pandas as pd
```

```python
url = "http://www.nuforc.org/webreports/ndxloc.html"
```

```python
content = requests.get(url)
```

```python
soup = BeautifulSoup(content.text, "html.parser")
```

```python
soup.prettify()
```

```python
#soup.find_all("a.href")
states=[]
for link in soup.find_all('a'):  # this will find all the hyper links
    # print type(link)  ## <class 'bs4.element.Tag'>
    #print link
    states.append(str(link.get_text()))
    


states = states[1:]

print states


```

```python
df = pd.read_csv("US_State_population2016.csv")
df['population'][0].replace (",", "",1) 


state_dict = {st[1:].lower():df['StateCode'].values[i] for i, st in enumerate(df['State'].values)}
state_dict
df['Country'] = ['USA' for i in range(56)]  # Add a new column: 'USA'.
```

```python
# Add some Canada columns (not all done yet)

df.loc[len(df)] = ['ON', 'ONTARIO,CAN', '13,448,494', 'CAN']

df.loc[len(df)] = ['AB', 'ALBERTA,CANADA', '4,067,175', 'CAN']

df.loc[len(df)] = ['BC', 'BRITISH COLUMBIA,CAN', '4,648,055', 'CAN']

df.loc[len(df)] = ['Manitoba', 'MANITOBA,CANADA', '1,278,365', 'CAN']

df.loc[len(df)] = ['NEW BRUNSWICK,CAN', 'NEW BRUNSWICK,CAN', '747,101', 'CAN']


df

```

```python
df['nsightings'] = [0 for n in range(len(df))]

for sight_info in state_count:
    #if df [df['State'] == "CALIFORNIA"]:
    pass
    
    try:
        statecode = state_dict[sight_info[0].lower()]
        #print statecode
        index = df.loc[df['StateCode']==statecode].index[0]
        #print index
        df.loc[index, 'nsightings'] = sight_info[1]
    except:
        #statecode = state_dict[sight_info[0]]
        #print statecode
        pass
```

```python
sight_rate = [(df['StateCode'][n], df['nsightings'][n], 
  int(df['population'][n].replace (",", "") ),
  1000*float(df['nsightings'][n]) / int(df['population'][n].replace (",", ""))  )

  for n in range(len(df))]
```

```python
# Sorted by the sighting rate
sorted_sight_rate = sorted(sight_rate, reverse=True, key=lambda x: x[3])
sorted_sight_rate
```
